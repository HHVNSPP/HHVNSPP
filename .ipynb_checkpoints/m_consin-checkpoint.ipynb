```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

#import warnings
import glob
```

```python
#warnings.filterwarnings('ignore')
plt.style.use('seaborn')
sns.set()
```

```python
data_s = pd.read_csv('sol_m_sin_4obj_64_1.txt.csv', sep=';', header=None)
data_ns = pd.read_csv('sol_m_sin_4obj_64_1 - copia.txt.csv', sep=';', header=None)
```

```python
data_s.columns = ['col_{}'.format(idx) for idx in range(1, len(data_s.columns) + 1)]
data_ns.columns = ['col_{}'.format(idx) for idx in range(1, len(data_ns.columns) + 1)]
```

```python
data_s.head()
```

```python
data_s.shape
data_ns.shape
```

```python
data_s.set_index('col_1', inplace=True)
data_ns.set_index('col_1', inplace=True)
```

```python
data_s.head()
```

```python
data_s = data_s.loc['2020-05-05':]
data_ns = data_ns.loc['2020-05-05':]
```

```python
data_s.head()
```

```python
summary_s = data_s.groupby('col_1')[['col_2', 'col_8', 'col_9', 'col_10', 'col_11']].max()
summary_ns = data_s.groupby('col_1')[['col_2', 'col_8', 'col_9', 'col_10', 'col_11']].max()
```

```python
summary_s.head()
```

```python
prop_s = data_s.groupby('col_1')['col_2'].agg(['mean', 'count']).groupby('mean').mean().reset_index().sort_values('count', ascending=False)
prop_ns = data_ns.groupby('col_1')['col_2'].agg(['mean', 'count']).groupby('mean').mean().reset_index().sort_values('count', ascending=False)
```

```python
prop_s.columns = ['time_elapsed', 'average_sol']
prop_ns.columns = ['time_elapsed', 'average_sol']
prop_s.to_csv("m128s.csv",index=None,header=None)
prop_ns.to_csv("m128ns.csv",index=None,header=None)
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
g = sns.lineplot(x='time_elapsed', y='average_sol', data=prop_s, markers=True)
_ = sns.scatterplot(x='time_elapsed', y='average_sol', data=prop_s, color='blue', markers='+')
h = sns.lineplot(x='time_elapsed', y='average_sol', data=prop_ns, markers=True)
plt.xscale('log')
a = sns.scatterplot(x='time_elapsed', y='average_sol', data=prop_ns, color='red', markers='+')
h.set_xlabel('Tiempo transcurrido en segundos')
a = g.set_ylabel('Promedio de soluciones no dominadas encontradas')
```
